# Images
Displaying, converting, manipulating and saving image data can be done using the `cv2` module. The `cv2` is a popular computer vision library for dealing with video and image data. Learning how to deal with image data can be useful in areas such as computer vision and machine learning (object detection/segmentation etc).

### Imports

```python
import cv2
from matplotlib import pyplot as plt
from IPython.display import Image
```

### Set image path
This is just a random image I found pixabay

```python
image_path = '../inputs/image-1465348_960_720.jpg'
```

```python
Image(image_path)
```

### Read in image 
Using `cv2.imread()` function we can read in an image as a multi dimensional array

```python
image = cv2.imread(image_path)
```

### Inspecting shape
The image is 960 by 642 pixels. It is in RGB mode because of the depth of 3.

```python
image.shape
```

### The structure
The image consist of 624 arrays, each of these arrays has 960 objects, also arrays. And in those arrays are the values of the pixels in BGR. cv2 reads in BGR mode by default, instead of RGB.

```python
len(image)
```

```python
len(image[0])
```

```python
image[0]
```

### Resizing images
You can resize images using a scale by calling the cv2.resize() function on an image, specifying your scale.

```python
scale = 0.6
smaller_image = cv2.resize(image, (0,0), fx = scale, fy = scale)
```

```python
smaller_image.shape
```

### Cropping images
Cropping, saving and displaying a cropped image by slicing the original image array

```python
cropped_image = image[120:320, 300:600]
cv2.imwrite('../outputs/cropped.png', cropped_image)
```

```python
Image('../outputs/cropped.png')
```

### Conversion to HSV (hue satuarion value) mode
The colour mode of an image can be converted, the resulting image can be saved using `cv2.imwrite()`

```python
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
cv2.imwrite('../outputs/hsv.png', hsv_image)
```

```python
Image('../outputs/hsv.png')
```

### Conversion to grey scale mode
You can remove colour using gray scale mode

```python
grey_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('../outputs/grey.png', grey_image)
```

```python
Image('../outputs/grey.png')
```
