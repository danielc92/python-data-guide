# Video
Conversion of video file to images and images to video file.

## Converting video to images

```python
import pandas
import cv2
import os
```

```python
video_path = '../inputs/illegal-tipping.mp4'
```

### Create `VideoCapture` object
From this we can do things like call the frames per second, total frames and read in frames one by one in sequential order

```python
stream = cv2.VideoCapture(video_path)
```

### Store the frame rate and total frame count
These are needed to process the video and to recreate the video later on

```python
frame_rate = stream.get(cv2.CAP_PROP_FPS)
print(frame_rate)
```

```python
frame_count = stream.get(cv2.CAP_PROP_FRAME_COUNT)
print(frame_count)
```

### Set image output folder

```python
images_output = '../outputs/video_images/'
```

### Process
While the frame counter is less or equal to the total frames available in the input video file, save each frame to a .png file using the `imwrite` function from `cv2`. Passing in the save path and frame as inputs.

```python
counter = 0

while(counter <= frame_count):
    
    ret, frame = stream.read()
    
    if ret:
        save_path = os.path.join(images_output, '{}.png'.format(counter))
        cv2.imwrite(save_path, frame)
        print('frame has been saved to {}'.format(save_path))
        
    counter = counter + 1
```

## Converting images to videos

```python
images_output = '../outputs/video_images/'
```

### Store the file info
The images may not be in sequential order, therefore you should store the creation time of the file (`os.stat.st_ctime`). 

```python
file_info = [(os.path.join(images_output, filename), 
          os.stat(os.path.join(images_output, filename)).st_ctime) for filename in os.listdir(images_output)]
```

### Store the file info in `pandas.DataFrame`
This just makes the data easier to view and sort

```python
data = pandas.DataFrame(file_info)
```

```python
data.columns = ['path', 'ctime']
```

### Sort by Creation Time
Ensures the images are in the same order they were created by `cv2.imwrite()`

```python
data = data.sort_values('ctime')
```

### Preview first 10 rows
Notice the index of the dataframe is scrambled. Indicating that the original order was not in sequential order.

```python
data.head(10)
```

### Store image objects
From the image paths in the dataframe, create a list of the image data itself. Each image is represented by a numpy nd array

```python
images = [cv2.imread(path) for path in list(data['path'])]
```

### Inspect the shape of one image
The `.shape` property shows (height, width, depth).
The `cv2.VideoWriter` object takes in shape argument but it is (WIDTH, HEIGHT)

```python
images[0].shape
```

```python
h, w, layers = images[0].shape
size = (w, h)
```

### Convert images to video
Set up a `cv2.VideoWriter` instance and being looping through the image data to write it to disk. call `.release()` at the end.

```python
output = cv2.VideoWriter('../outputs/output.avi', cv2.VideoWriter_fourcc(*'DIVX'), frame_rate, size, True)

for i in range(len(images)):
    print('Writing {} / {}'.format(i, len(images)))
    output.write(images[i])

output.release()
```
